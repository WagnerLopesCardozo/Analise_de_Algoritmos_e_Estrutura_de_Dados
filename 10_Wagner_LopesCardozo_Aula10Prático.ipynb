#<font color=lightgreen>Implementação do algoritmo para função Hash que faz resolução de colisões por encadeamento com a função Hash da divisão.

```python
class HashTable:
    def __init__(self, size):
        self.size = size  # Tamanho da tabela hash
        self.table = [[] for _ in range(size)]  # Inicializa a tabela com listas vazias para encadeamento

    def hash_function(self, key):
        """Função hash usando o método da divisão."""
        return key % self.size

    def insert(self, key):
        """Insere uma chave na tabela hash."""
        index = self.hash_function(key)
        # Verifica se a chave já está presente na lista encadeada
        if key not in self.table[index]:
            self.table[index].append(key)  # Adiciona a chave na lista encadeada
            print(f'Chave {key} inserida na posição {index}.')
        else:
            print(f'Chave {key} já está presente na posição {index}.')

    def search(self, key):
        """Busca uma chave na tabela hash."""
        index = self.hash_function(key)
        if key in self.table[index]:
            print(f'Chave {key} encontrada na posição {index}.')
            return True
        else:
            print(f'Chave {key} não encontrada.')
            return False

    def display(self):
        """Exibe a tabela hash com suas listas encadeadas."""
        for index, chain in enumerate(self.table):
            print(f'Posição {index}: {chain}')


# Exemplo de uso
if __name__ == "__main__":
    hash_table = HashTable(9)  # Cria uma tabela hash de tamanho 9

    # Chaves para inserir
    keys = [5, 28, 19, 15, 20, 33, 12, 17, 10]

    # Inserindo chaves na tabela hash
    for key in keys:
        hash_table.insert(key)

    # Exibindo a tabela hash
    print("\nTabela Hash:")
    hash_table.display()

    # Buscando algumas chaves
    hash_table.search(15)
    hash_table.search(30)
```

#<font color=lightgreen>O mesmo algoritmo acima, porém com a exibição da tabela Hash formatado em forma de tabela para função Hash que faz resolução de colisões por encadeamento com a função Hash da divisão.

```python
from tabulate import tabulate

class HashTable:
    def __init__(self, size):
        self.size = size  # Tamanho da tabela hash
        self.table = [[] for _ in range(size)]  # Inicializa a tabela com listas vazias para encadeamento

    def hash_function(self, key):
        """Função hash usando o método da divisão."""
        return key % self.size

    def insert(self, key):
        """Insere uma chave na tabela hash."""
        index = self.hash_function(key)
        # Verifica se a chave já está presente na lista encadeada
        if key not in self.table[index]:
            self.table[index].append(key)  # Adiciona a chave na lista encadeada
            print(f'Chave {key} inserida na posição {index}.')
        else:
            print(f'Chave {key} já está presente na posição {index}.')

    def search(self, key):
        """Busca uma chave na tabela hash."""
        index = self.hash_function(key)
        if key in self.table[index]:
            print(f'Chave {key} encontrada na posição {index}.')
            return True
        else:
            print(f'Chave {key} não encontrada.')
            return False

    def display(self):
        """Exibe a tabela hash com suas listas encadeadas usando tabulate."""
        # Prepara os dados para exibição
        display_data = []
        for index, chain in enumerate(self.table):
            display_data.append([index, ', '.join(map(str, chain))])  # Formata a lista encadeada como string

        # Usa tabulate para exibir a tabela
        print(tabulate(display_data, headers=['Posição', 'Chaves'], tablefmt='pretty'))


# Exemplo de uso
if __name__ == "__main__":
    hash_table = HashTable(9)  # Cria uma tabela hash de tamanho 9

    # Chaves para inserir
    keys = [5, 28, 19, 15, 20, 33, 12, 17, 10]

    # Inserindo chaves na tabela hash
    for key in keys:
        hash_table.insert(key)

    # Exibindo a tabela hash
    print("\nTabela Hash:")
    hash_table.display()

    # Buscando algumas chaves
    hash_table.search(15)
    hash_table.search(30)
```

#<font color=lightgreen>Implementação do algoritmo para função Hash que faz resolução de colisões por endereçamento aberto usando sondagem linear.

```python
class LinearProbingHashTable:
    def __init__(self, size):
        self.size = size  # Tamanho da tabela hash
        self.table = [None] * size  # Inicializa a tabela com None

    def hash_function(self, key):
        """Função hash usando o método da divisão."""
        return key % self.size

    def insert(self, key):
        """Insere uma chave na tabela hash usando sondagem linear."""
        index = self.hash_function(key)

        # Sondagem linear para encontrar uma posição vazia
        while self.table[index] is not None:
            if self.table[index] == key:  # Verifica se a chave já está presente
                print(f'Chave {key} já está presente na posição {index}.')
                return
            index = (index + 1) % self.size  # Move para a próxima posição

        self.table[index] = key  # Insere a chave na tabela
        print(f'Chave {key} inserida na posição {index}.')

    def search(self, key):
        """Busca uma chave na tabela hash usando sondagem linear."""
        index = self.hash_function(key)

        # Sondagem linear para encontrar a chave
        while self.table[index] is not None:
            if self.table[index] == key:
                print(f'Chave {key} encontrada na posição {index}.')
                return True
            index = (index + 1) % self.size  # Move para a próxima posição

        print(f'Chave {key} não encontrada.')
        return False

    def display(self):
        """Exibe a tabela hash usando tabulate."""
        display_data = []
        for index, value in enumerate(self.table):
            display_data.append([index, value])  # Adiciona o índice e o valor na tabela

        # Usa tabulate para exibir a tabela
        print(tabulate(display_data, headers=['Posição', 'Chave'], tablefmt='pretty'))


# Exemplo de uso
if __name__ == "__main__":
    hash_table = LinearProbingHashTable(11)  # Cria uma tabela hash de tamanho 11

    # Chaves para inserir
    keys = [10, 22, 31, 4, 15, 28, 17, 88, 59]

    # Inserindo chaves na tabela hash
    for key in keys:
        hash_table.insert(key)

    # Exibindo a tabela hash
    print("\nTabela Hash:")
    hash_table.display()

    # Buscando algumas chaves
    hash_table.search(15)
    hash_table.search(30)
```

#<font color=lightgreen>Algoritmo que insere 100.000 elementos gerados aleatoriamente no intervalo de 0 a 100.000 em uma tabela hash com m = 50.000 posições no encadeamento e m = 100.000 no endereçamento aberto e calcula e exibe na tela o tempo de inserção em cada caso. Depois pesquisa o elemento 10.000

```python
import numpy as np
import time
from tabulate import tabulate

class ChainedHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return key % self.size

    def insert(self, key):
        index = self.hash_function(key)
        if key not in self.table[index]:
            self.table[index].append(key)

    def search(self, key):
        index = self.hash_function(key)
        return key in self.table[index]

class LinearProbingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return key % self.size

    def insert(self, key):
        index = self.hash_function(key)
        while self.table[index] is not None:
            if self.table[index] == key:
                return  # A chave já está presente
            index = (index + 1) % self.size
        self.table[index] = key

    def search(self, key):
        index = self.hash_function(key)
        while self.table[index] is not None:
            if self.table[index] == key:
                return True
            index = (index + 1) % self.size
        return False

# Função para medir o tempo de inserção e busca
def measure_time():
    # Gerando 100.000 elementos aleatórios
    random_elements = np.random.randint(0, 100001, 100000).tolist()

    # Encadeamento
    chained_hash_table = ChainedHashTable(50000)
    start_time = time.time()
    for elem in random_elements:
        chained_hash_table.insert(elem)
    insertion_time_chained = time.time() - start_time

    start_time = time.time()
    search_result_chained = chained_hash_table.search(10000)
    search_time_chained = time.time() - start_time

    # Endereçamento aberto
    linear_probing_hash_table = LinearProbingHashTable(100000)
    start_time = time.time()
    for elem in random_elements:
        linear_probing_hash_table.insert(elem)
    insertion_time_linear = time.time() - start_time

    start_time = time.time()
    search_result_linear = linear_probing_hash_table.search(10000)
    search_time_linear = time.time() - start_time

    # Exibindo resultados com a formatação especificada
    results = [
        ["Hash por encadeamento", "Hash por endereçamento aberto"],
        [f"Tempo Inserção\nFunção hash\ndivisão: {insertion_time_chained:.6f} s",
         f"Tempo Inserção\nSondagem Linear: {insertion_time_linear:.6f} s"],
        [f"Tempo Busca\nFunção hash\ndivisão: {search_time_chained:.6f} s",
         f"Tempo Busca\nSondagem Linear: {search_time_linear:.6f} s"],
    ]

    print(tabulate(results, tablefmt='grid'))

# Executando a medição de tempo
measure_time()
```

```python
import numpy as np
import time
from tabulate import tabulate

class ChainedHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return key % self.size

    def insert(self, key):
        index = self.hash_function(key)
        if key not in self.table[index]:
            self.table[index].append(key)

    def search(self, key):
        index = self.hash_function(key)
        return key in self.table[index]

class LinearProbingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return key % self.size

    def insert(self, key):
        index = self.hash_function(key)
        while self.table[index] is not None:
            if self.table[index] == key:
                return  # A chave já está presente
            index = (index + 1) % self.size
        self.table[index] = key

    def search(self, key):
        index = self.hash_function(key)
        while self.table[index] is not None:
            if self.table[index] == key:
                return True
            index = (index + 1) % self.size
        return False

# Função para medir o tempo de inserção e busca
def measure_time(iterations=50):
    total_insertion_time_chained = 0
    total_search_time_chained = 0
    total_insertion_time_linear = 0
    total_search_time_linear = 0

    for _ in range(iterations):
        # Gerando 100.000 elementos aleatórios
        random_elements = np.random.randint(0, 100001, 100000).tolist()

        # Encadeamento
        chained_hash_table = ChainedHashTable(50000)
        start_time = time.time()
        for elem in random_elements:
            chained_hash_table.insert(elem)
        total_insertion_time_chained += time.time() - start_time

        start_time = time.time()
        search_result_chained = chained_hash_table.search(10000)
        total_search_time_chained += time.time() - start_time

        # Endereçamento aberto
        linear_probing_hash_table = LinearProbingHashTable(100000)
        start_time = time.time()
        for elem in random_elements:
            linear_probing_hash_table.insert(elem)
        total_insertion_time_linear += time.time() - start_time

        start_time = time.time()
        search_result_linear = linear_probing_hash_table.search(10000)
        total_search_time_linear += time.time() - start_time

    # Calculando as médias
    avg_insertion_time_chained = total_insertion_time_chained / iterations
    avg_search_time_chained = total_search_time_chained / iterations
    avg_insertion_time_linear = total_insertion_time_linear / iterations
    avg_search_time_linear = total_search_time_linear / iterations

    # Exibindo resultados com a formatação especificada
    print()
    print('Média Aritimética Para 50 Execuções:')
    print()
    results = [
        ["Hash por encadeamento", "Hash por endereçamento aberto"],
        [f"Tempo Inserção\nFunção hash\ndivisão: {avg_insertion_time_chained:.6f} s",
         f"Tempo Inserção\nSondagem Linear: {avg_insertion_time_linear:.6f} s"],
        [f"Tempo Busca\nFunção hash\ndivisão: {avg_search_time_chained:.6f} s",
         f"Tempo Busca\nSondagem Linear: {avg_search_time_linear:.6f} s"],
    ]

    print(tabulate(results, tablefmt='grid'))

# Executando a medição de tempo
measure_time(50)
```

#<font color=lightgreen>Algoritmo que implementa o método de Tabela Hash por Encadeamento e por Endereçamento Aberto com 100 nomes, sendo 50 femininos e 50 masculinos com seus respectivos gêneros e idades, mulheres entre 35 à 75 anos e homens entre 40 à 80 anos, com exames de mamografia para as pacientes femininas e toque retal para os pacientes masculinos. O algoritmo calcula o tempo de inserção e busca nos dois métodos implementados para as Tabelas Hash's.

```python
import numpy as np
import time
from tabulate import tabulate

# Listas de nomes típicos para identificação de gênero
female_names = ["Ana", "Maria", "Clara", "Beatriz", "Julia", "Camila", "Larissa", "Patricia", "Fernanda", "Luciana"]
male_names = ["João", "Carlos", "Pedro", "Lucas", "Mateus", "Gabriel", "Ricardo", "Paulo", "José", "Daniel"]

# Funções para exibir texto colorido usando códigos ANSI
def color_text(text, color_code):
    return f"\033[{color_code}m{text}\033[0m"

def color_mamografia():
    return color_text("Mamografia", "35")  # Magenta (rosa)

def color_toque_retal():
    return color_text("Toque Retal", "34")  # Azul

class ChainedHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key):
        index = self.hash_function(key)
        if key not in self.table[index]:
            self.table[index].append(key)

    def search(self, key):
        index = self.hash_function(key)
        return key in self.table[index]

class LinearProbingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key):
        index = self.hash_function(key)
        while self.table[index] is not None:
            if self.table[index] == key:
                return  # A chave já está presente
            index = (index + 1) % self.size
        self.table[index] = key

    def search(self, key):
        index = self.hash_function(key)
        while self.table[index] is not None:
            if self.table[index] == key:
                return True
            index = (index + 1) % self.size
        return False

# Função para medir o tempo de inserção e busca
def measure_time():
    # Gerando 50 nomes de pacientes femininos e 50 masculinos com idades dentro dos intervalos especificados
    female_patients = [(np.random.choice(female_names), "Feminino", np.random.randint(35, 76)) for _ in range(50)]
    male_patients = [(np.random.choice(male_names), "Masculino", np.random.randint(40, 81)) for _ in range(50)]

    # Lista completa de pacientes com gênero e idade
    patients = female_patients + male_patients

    # Embaralhando a lista de pacientes
    np.random.shuffle(patients)

    # Selecionando 10 pacientes femininos e 10 masculinos para busca
    female_search_patients = np.random.choice(range(len(female_patients)), 10, replace=False)
    male_search_patients = np.random.choice(range(len(male_patients)), 10, replace=False)

    # Inicializando tabelas hash
    chained_hash_table = ChainedHashTable(5000)
    linear_probing_hash_table = LinearProbingHashTable(5000)

    # Medindo tempo de inserção para encadeamento
    start_time = time.time()
    for patient in patients:
        chained_hash_table.insert(patient[0])  # Inserindo apenas o nome do paciente
    insertion_time_chained = time.time() - start_time

    # Medindo tempo de busca para encadeamento (feminino e masculino)
    start_time = time.time()
    for idx in female_search_patients:
        chained_hash_table.search(female_patients[idx][0])  # Buscando apenas o nome do paciente
    for idx in male_search_patients:
        chained_hash_table.search(male_patients[idx][0])  # Buscando apenas o nome do paciente
    search_time_chained = time.time() - start_time

    # Medindo tempo de inserção para endereçamento aberto
    start_time = time.time()
    for patient in patients:
        linear_probing_hash_table.insert(patient[0])  # Inserindo apenas o nome do paciente
    insertion_time_linear = time.time() - start_time

    # Medindo tempo de busca para endereçamento aberto (feminino e masculino)
    start_time = time.time()
    for idx in female_search_patients:
        linear_probing_hash_table.search(female_patients[idx][0])  # Buscando apenas o nome do paciente
    for idx in male_search_patients:
        linear_probing_hash_table.search(male_patients[idx][0])  # Buscando apenas o nome do paciente
    search_time_linear = time.time() - start_time

    # Exibindo tabelas com os resultados
    results = [
        ["Método", "Tempo Inserção", "Tempo Busca"],
        ["Hash por encadeamento", f"{insertion_time_chained:.6f} s", f"{search_time_chained:.6f} s"],
        ["Hash por endereçamento aberto", f"{insertion_time_linear:.6f} s", f"{search_time_linear:.6f} s"],
    ]

    print("Tabela de Resultados:")
    print(tabulate(results, headers="firstrow", tablefmt='grid'))

    # Exibindo a lista de pacientes embaralhados
    print("\nTabela de Pacientes Criados (com Exames e Idades):")
    patient_results = []
    for i, (name, gender, age) in enumerate(patients):
        exam = color_mamografia() if gender == "Feminino" else color_toque_retal()
        patient_results.append([i + 1, name, gender, age, exam])
    print(tabulate(patient_results, headers=["#", "Nome do Paciente", "Gênero", "Idade", "Exame"], tablefmt='grid'))

    # Tabelas separadas para pacientes femininos e masculinos buscados
    print("\nTabela de Pacientes Femininos Buscados:")
    female_results = [[i + 1, female_patients[idx][0], female_patients[idx][1], female_patients[idx][2], color_mamografia()] for i, idx in enumerate(female_search_patients)]
    print(tabulate(female_results, headers=["#", "Nome do Paciente", "Gênero", "Idade", "Exame"], tablefmt='grid'))

    print("\nTabela de Pacientes Masculinos Buscados:")
    male_results = [[i + 1, male_patients[idx][0], male_patients[idx][1], male_patients[idx][2], color_toque_retal()] for i, idx in enumerate(male_search_patients)]
    print(tabulate(male_results, headers=["#", "Nome do Paciente", "Gênero", "Idade", "Exame"], tablefmt='grid'))

# Executando a medição de tempo
measure_time()
```
