#<font color=lightgreen>Algoritmo 1 (Com Uso da Biblioteca InterTools)

##Algoritmo que gera todas as permutações de n strings informadas pelo usuário, em ordem lexicográfica crescente, exibe na tela os resultados e também o tempo de execução do algoritmo.

image.png

```python
import itertools
import time

def generate_permutations(strings):
    for s in strings:
        permutations = sorted(set(itertools.permutations(s)))
        for perm in permutations:
            print(''.join(perm))
        print()

def main():
    # Leitura do número de strings
    n = int(input("Digite o número de strings: "))

    # Leitura das strings
    strings = []
    for _ in range(n):
        strings.append(input())

    # Imprime as entradas
    print("\nEntradas:")
    for s in strings:
        print(s)

    # Mede o tempo de execução
    start_time = time.time()

    # Gera e imprime as permutações
    print("\nSaídas:")
    generate_permutations(strings)

    end_time = time.time()
    execution_time = end_time - start_time
    print(f"Tempo de execução: {execution_time:.6f} segundos")

# Executa o código principal
if __name__ == "__main__":
    main()
```

#<font color=lightgreen>Algoritmo 2 (Implementado sem o uso da biblioteca InterTools)

##Este algoritmo processa e exibe o mesmo que o algoritmo acima, porém foi implementado sem o uso da biblioteca InterTools para que se possa fazer a análise assintótica.

```python
import time

def next_permutation(s):
    # Encontra o índice i tal que s[i] < s[i + 1]
    i = len(s) - 2
    while i >= 0 and s[i] >= s[i + 1]:               #Primeiro loop - O(n), sendo n o tamanho da string.
        i -= 1

    if i == -1:
        return False

    # Encontra o índice j tal que s[i] < s[j]
    j = len(s) - 1
    while s[j] <= s[i]:                                 #Segundo loop - O(n).
        j -= 1

    # Troca os elementos em i e j
    s[i], s[j] = s[j], s[i]                      #Trocar elementos tem complexidade O(1).

    # Reverte a sequência após o índice i
    s[i + 1:] = reversed(s[i + 1:])             #reverter a sublista cria uma complexidade O(n), por exemplo O(x) onde x seria o número de elementos na sublista.

    return True

def generate_permutations(s):
    # Ordena a string para começar com a permutação lexicograficamente menor
    s = sorted(s)                                                                       #Usando um raciocínio matemático algébrico:
                                                                                        #O(n)+O(n)+O(1)+O(n)=O(3n+1)
    # Imprime a primeira permutação                                                     #logo a complexidade final será O(n)
    print(''.join(s))

    # Gera e imprime as próximas permutações até não ser possível mais permutar
    while next_permutation(s):
        print(''.join(s))

def main():
    # Leitura do número de strings
    n = int(input("Digite o número de strings: "))

    # Leitura das strings
    strings = []
    for _ in range(n):
        strings.append(input())

    # Imprime as entradas
    print("\nEntradas:")
    for s in strings:
        print(s)

    # Mede o tempo de execução
    start_time = time.time()

    # Gera e imprime as permutações
    print("\nSaídas:")
    for s in strings:
        generate_permutations(s)
        print()  # Linha em branco após cada grupo de permutações

    end_time = time.time()
    execution_time = end_time - start_time
    print(f"Tempo de execução: {execution_time:.6f} segundos")

# Executa o código principal
if __name__ == "__main__":
    main()
```
