#<font color=lightgreen>Estudo 1 - Recursivo

##Algoritimo com a fun√ß√£o recursiva implementada, por√©m este algoritmo por ser recursivo tem uma profundidade de busca de 966 vezes dentro da fun√ß√£o recursiva no vetor de 100.000 n√∫meros, sendo uma limita√ß√£o do Python. Al√©m do tempo de execu√ß√£o ser muito alto devido ao seu custo assint√≥tico Big(O) ser de O(n¬≤) representando uma complexidade quadr√°tica.

##Solu√ß√£o:

## Implementar o mesmo algoritimo de forma interativa.

```python
import random
import time
import math

# Fun√ß√£o recursiva para verificar se um n√∫mero √© primo
def is_prime(n, divisor=None):
    if n < 2:
        return False
    if divisor is None:
        divisor = n - 1
    if divisor == 1:
        return True
    if n % divisor == 0:
        return False
    return is_prime(n, divisor - 1)

# Fun√ß√£o para encontrar o maior n√∫mero primo em um vetor
def find_largest_prime(arr):
    max_prime = None
    for num in arr:
        if is_prime(num):
            if max_prime is None or num > max_prime:
                max_prime = num
    return max_prime

# Gerando um vetor aleat√≥rio de 100.000 elementos
arr = [random.randint(1, 966) for _ in range(100000)]

# Medindo o tempo de execu√ß√£o
start_time = time.time()
largest_prime = find_largest_prime(arr)
end_time = time.time()

# Calculando o tempo de execu√ß√£o
execution_time = end_time - start_time

# An√°lise de Complexidade
# A fun√ß√£o is_prime tem complexidade O(n) no pior caso, onde n √© o n√∫mero sendo testado.
# A fun√ß√£o find_largest_prime tem complexidade O(m * n), onde m √© o n√∫mero de elementos no vetor e n √© o valor m√©dio dos n√∫meros no vetor.

complexity_analysis = """
An√°lise de Complexidade:
- A fun√ß√£o is_prime tem complexidade O(n), onde n √© o n√∫mero sendo testado.
- A fun√ß√£o find_largest_prime, que verifica cada n√∫mero no vetor, tem complexidade O(m * n),
  onde m √© o tamanho do vetor e n √© o valor m√©dio dos n√∫meros no vetor.

Portanto, a complexidade total √© O(m * n), que √© aproximadamente O(n^2) para grandes n,
considerando que m = n (n√∫meros no vetor e n√∫mero sendo testado).
Esta √© uma complexidade quadr√°tica.
"""

# Exibindo os resultados
print(f"O maior n√∫mero primo encontrado foi: {largest_prime}")
print(f"Tempo de execu√ß√£o: {execution_time:.6f} segundos")
print(complexity_analysis)
```

#<font color=lightgreen>Estudo 2 - Interativo

##O mesmo algoritimo com a fun√ß√£o interativa implementada, agora este algoritmo pode ser executado com uma profundidade de 1.000.000 pecorrendo todo o vetor de 100.000 n√∫meros.Al√©m do tempo de execu√ß√£o ser bem mais baixo devido ao seu custo assint√≥tico Big(O) ser de O(m*sqrt(n)) representando uma complexidade denominada sublinear.

#Defini√ß√£o

Um algoritmo de complexidade sublinear √© aquele cuja complexidade de tempo cresce mais lentamente do que o tamanho da entrada. Isso significa que, √† medida que o tamanho da entrada aumenta, o tempo que o algoritmo leva para ser executado aumenta a uma taxa mais lenta.

Exemplos de complexidades sublineares incluem:

* O(log n): Busca bin√°ria em uma lista ordenada.

* O(sqrt(n)): Algoritmos que iteram sobre a raiz quadrada do tamanho da entrada.

Esses algoritmos s√£o geralmente muito eficientes, pois podem lidar com grandes entradas em um tempo relativamente curto. Eles s√£o frequentemente usados em tarefas como pesquisa, classifica√ß√£o e processamento de grandes conjuntos de dados.

* https://stackoverflow.com/questions/32311924/what-are-sublinear-algorithms

```python
import random
import time
import math

# Fun√ß√£o iterativa para verificar se um n√∫mero √© primo
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Fun√ß√£o para encontrar o maior n√∫mero primo em um vetor
def find_largest_prime(arr):
    max_prime = None
    for num in arr:
        if is_prime(num):
            if max_prime is None or num > max_prime:
                max_prime = num
    return max_prime

# Gerando um vetor aleat√≥rio de 100.000 elementos
arr = [random.randint(1, 1000000) for _ in range(100000)]

# Medindo o tempo de execu√ß√£o
start_time = time.time()
largest_prime = find_largest_prime(arr)
end_time = time.time()

# Calculando o tempo de execu√ß√£o
execution_time = end_time - start_time

# An√°lise de Complexidade
# A fun√ß√£o is_prime tem complexidade O(‚àön) no pior caso, onde n √© o n√∫mero sendo testado.
# A fun√ß√£o find_largest_prime tem complexidade O(m * ‚àön), onde m √© o n√∫mero de elementos no vetor e n √© o valor m√©dio dos n√∫meros no vetor.

complexity_analysis = """
An√°lise de Complexidade:
- A fun√ß√£o is_prime tem complexidade O(‚àön), onde n √© o n√∫mero sendo testado.
- A fun√ß√£o find_largest_prime, que verifica cada n√∫mero no vetor, tem complexidade O(m * ‚àön),
  onde m √© o tamanho do vetor e n √© o valor m√©dio dos n√∫meros no vetor.

Portanto, a complexidade total √© O(m * ‚àön), que √© mais eficiente do que O(n^2) para grandes n.
Esta √© uma complexidade sublinear.
"""

# Exibindo os resultados
print(f"O maior n√∫mero primo encontrado foi: {largest_prime}")
print(f"Tempo de execu√ß√£o: {execution_time:.6f} segundos")
print(complexity_analysis)
```

```python
import random
import time
import math
import matplotlib.pyplot as plt

# Fun√ß√£o iterativa para verificar se um n√∫mero √© primo
def is_prime_iterative(n):
    # Verifica se n √© menor ou igual a 1, pois n√∫meros menores ou iguais a 1 n√£o s√£o primos
    if n <= 1:
        return False
    # Verifica se n √© 2 ou 3, pois esses n√∫meros s√£o primos
    if n <= 3:
        return True
    # Elimina m√∫ltiplos de 2 e 3 que n√£o s√£o primos (exceto o pr√≥prio 2 e 3)
    if n % 2 == 0 or n % 3 == 0:
        return False
    # Inicia o divisor i em 5 e incrementa 6 a cada passo para eliminar m√∫ltiplos de 2 e 3
    i = 5
    while i * i <= n:
        # Se n for divis√≠vel por i ou i + 2, n√£o √© primo
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    # Se passar por todas as verifica√ß√µes, n √© primo
    return True

# Fun√ß√£o recursiva para verificar se um n√∫mero √© primo
def is_prime_recursive(n, divisor=None):
    # N√∫meros menores que 2 n√£o s√£o primos
    if n < 2:
        return False
    # Se o divisor n√£o for especificado, inicializa com n - 1
    if divisor is None:
        divisor = n - 1
    # Se o divisor for 1, n √© primo
    if divisor == 1:
        return True
    # Se n for divis√≠vel por divisor, n√£o √© primo
    if n % divisor == 0:
        return False
    # Chama a fun√ß√£o recursivamente, reduzindo o divisor
    return is_prime_recursive(n, divisor - 1)

# Fun√ß√£o para encontrar o maior n√∫mero primo em um vetor
def find_largest_prime(arr, prime_func):
    max_prime = None  # Inicializa a vari√°vel para armazenar o maior n√∫mero primo
    # Itera sobre cada n√∫mero no vetor
    for num in arr:
        # Verifica se o n√∫mero √© primo usando a fun√ß√£o fornecida (iterativa ou recursiva)
        if prime_func(num):
            # Se max_prime for None ou o n√∫mero atual for maior que max_prime, atualiza max_prime
            if max_prime is None or num > max_prime:
                max_prime = num
    # Retorna o maior n√∫mero primo encontrado
    return max_prime

# Gerando um vetor aleat√≥rio de 100.000 elementos, com valores entre 1 e 100
arr = [random.randint(1, 960) for _ in range(100000)]

# Medindo o tempo de execu√ß√£o para o m√©todo iterativo
start_time_iterative = time.time()  # Inicia a contagem do tempo
largest_prime_iterative = find_largest_prime(arr, is_prime_iterative)  # Encontra o maior primo usando a fun√ß√£o iterativa
end_time_iterative = time.time()  # Finaliza a contagem do tempo
execution_time_iterative = end_time_iterative - start_time_iterative  # Calcula o tempo de execu√ß√£o

# Medindo o tempo de execu√ß√£o para o m√©todo recursivo
start_time_recursive = time.time()  # Inicia a contagem do tempo
largest_prime_recursive = find_largest_prime(arr, is_prime_recursive)  # Encontra o maior primo usando a fun√ß√£o recursiva
end_time_recursive = time.time()  # Finaliza a contagem do tempo
execution_time_recursive = end_time_recursive - start_time_recursive  # Calcula o tempo de execu√ß√£o

# Exibindo os resultados
print(f"Maior n√∫mero primo encontrado (Iterativo): {largest_prime_iterative}")
print(f"Tempo de execu√ß√£o (Iterativo): {execution_time_iterative:.6f} segundos")
print(f"Maior n√∫mero primo encontrado (Recursivo): {largest_prime_recursive}")
print(f"Tempo de execu√ß√£o (Recursivo): {execution_time_recursive:.6f} segundos")

# Plotando os gr√°ficos de compara√ß√£o de tempo de execu√ß√£o e da an√°lise assint√≥tica (Big-O)
plt.figure(figsize=(14, 6))

# Gr√°fico de tempo de execu√ß√£o
plt.subplot(1, 2, 1)
plt.bar(['Iterativo', 'Recursivo'], [execution_time_iterative, execution_time_recursive], color=['blue', 'red'])
plt.ylabel('Tempo de Execu√ß√£o (s)')
plt.title('Compara√ß√£o de Tempo de Execu√ß√£o: M√©todo Iterativo vs Recursivo')

# Gr√°fico da an√°lise assint√≥tica (Big-O)
n_values = list(range(1, 101))
iterative_big_o = [math.sqrt(n) for n in n_values]  # Estimativa O(‚àön) para o iterativo
recursive_big_o = n_values  # O(n) para o recursivo

plt.subplot(1, 2, 2)
plt.plot(n_values, iterative_big_o, label="Iterativo O(‚àön)", color='blue', marker='o')
plt.plot(n_values, recursive_big_o, label="Recursivo O(n)", color='red', marker='o')
plt.xlabel('Valor de n')
plt.ylabel('Fun√ß√£o Assint√≥tica')
plt.title('An√°lise Assint√≥tica (Big-O)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import os
import psutil
import platform
import subprocess
import tensorflow as tf
import torch

def get_cpu_info():
    cpu_info = {}
    cpu_info['Name'] = platform.processor()
    cpu_info['Cores'] = psutil.cpu_count(logical=False)
    cpu_info['Threads'] = psutil.cpu_count(logical=True)
    return cpu_info

def get_ram_info():
    ram_info = psutil.virtual_memory()
    return {'Total RAM (GB)': ram_info.total / (1024 ** 3),
            'Available RAM (GB)': ram_info.available / (1024 ** 3)}

def get_disk_info():
    disk_info = psutil.disk_usage('/')
    return {'Total Storage (GB)': disk_info.total / (1024 ** 3),
            'Used Storage (GB)': disk_info.used / (1024 ** 3),
            'Free Storage (GB)': disk_info.free / (1024 ** 3)}

def get_gpu_info():
    gpu_info = {}
    if torch.cuda.is_available():
        gpu_info['CUDA available'] = True
        gpu_info['GPU Name'] = torch.cuda.get_device_name(0)
        gpu_info['CUDA Version'] = torch.version.cuda
        gpu_info['Number of GPUs'] = torch.cuda.device_count()
    else:
        gpu_info['CUDA available'] = False
    return gpu_info

def get_python_info():
    python_info = {}
    python_info['Python Version'] = platform.python_version()
    python_info['TensorFlow Version'] = tf.__version__
    python_info['PyTorch Version'] = torch.__version__
    return python_info

def get_system_info():
    system_info = {}
    system_info['OS'] = platform.system()
    system_info['OS Version'] = platform.version()
    system_info['Machine'] = platform.machine()
    return system_info

# Extraindo todas as informa√ß√µes
cpu_info = get_cpu_info()
ram_info = get_ram_info()
disk_info = get_disk_info()
gpu_info = get_gpu_info()
python_info = get_python_info()
system_info = get_system_info()

# Exibindo o relat√≥rio completo
print("==== Relat√≥rio do Ambiente de Desenvolvimento Integrado (IDE) ====\n")
print("=== Informa√ß√µes do Sistema ===")
for key, value in system_info.items():
    print(f"{key}: {value}")

print("\n=== Informa√ß√µes do Processador ===")
for key, value in cpu_info.items():
    print(f"{key}: {value}")

print("\n=== Informa√ß√µes de Mem√≥ria RAM ===")
for key, value in ram_info.items():
    print(f"{key}: {value:.2f} GB")

print("\n=== Informa√ß√µes de Armazenamento ===")
for key, value in disk_info.items():
    print(f"{key}: {value:.2f} GB")

print("\n=== Informa√ß√µes de GPU ===")
for key, value in gpu_info.items():
    if isinstance(value, bool):
        print(f"{key}: {'Sim' if value else 'N√£o'}")
    else:
        print(f"{key}: {value}")

print("\n=== Vers√£o do Python e Principais Bibliotecas ===")
for key, value in python_info.items():
    print(f"{key}: {value}")
```

```python
import os
import psutil
import platform
import subprocess
import torch
import tensorflow as tf
import pkg_resources

def get_cpu_info():
    cpu_info = {}
    cpu_info['Processor Name'] = platform.processor()
    cpu_info['Machine Type'] = platform.machine()
    cpu_info['Architecture'] = platform.architecture()[0]
    cpu_info['Physical Cores'] = psutil.cpu_count(logical=False)
    cpu_info['Logical Cores'] = psutil.cpu_count(logical=True)

    # Execu√ß√£o de comandos do shell para obter mais informa√ß√µes do processador
    try:
        cpu_info['Model Name'] = subprocess.check_output("lscpu | grep 'Model name'", shell=True).decode().strip().split(':')[1].strip()
        cpu_info['Vendor ID'] = subprocess.check_output("lscpu | grep 'Vendor ID'", shell=True).decode().strip().split(':')[1].strip()
        cpu_info['CPU MHz'] = subprocess.check_output("lscpu | grep 'CPU MHz'", shell=True).decode().strip().split(':')[1].strip()
    except Exception as e:
        cpu_info['Model Name'] = 'N/A'
        cpu_info['Vendor ID'] = 'N/A'
        cpu_info['CPU MHz'] = 'N/A'

    return cpu_info

def get_ram_info():
    ram_info = psutil.virtual_memory()
    return {
        'Total RAM (GB)': ram_info.total / (1024 ** 3),
        'Available RAM (GB)': ram_info.available / (1024 ** 3),
        'Used RAM (GB)': ram_info.used / (1024 ** 3),
        'RAM Usage (%)': ram_info.percent
    }

def get_disk_info():
    disk_info = psutil.disk_usage('/')
    return {
        'Total Storage (GB)': disk_info.total / (1024 ** 3),
        'Used Storage (GB)': disk_info.used / (1024 ** 3),
        'Free Storage (GB)': disk_info.free / (1024 ** 3),
        'Storage Usage (%)': disk_info.percent
    }

def get_gpu_info():
    gpu_info = {}
    if torch.cuda.is_available():
        gpu_info['CUDA available'] = True
        gpu_info['GPU Name'] = torch.cuda.get_device_name(0)
        gpu_info['CUDA Version'] = torch.version.cuda
        gpu_info['Number of GPUs'] = torch.cuda.device_count()
        gpu_info['GPU Memory Allocated (MB)'] = torch.cuda.memory_allocated(0) / (1024 ** 2)
        gpu_info['GPU Memory Reserved (MB)'] = torch.cuda.memory_reserved(0) / (1024 ** 2)
    else:
        gpu_info['CUDA available'] = False
        try:
            # Detalhes da GPU usando nvidia-smi
            gpu_info['GPU Details'] = subprocess.check_output("nvidia-smi", shell=True).decode()
        except:
            gpu_info['GPU Details'] = 'No GPU or not connected to a GPU runtime'
    return gpu_info

def get_python_info():
    python_info = {}
    python_info['Python Version'] = platform.python_version()
    python_info['TensorFlow Version'] = tf.__version__
    python_info['PyTorch Version'] = torch.__version__

    # Listar todas as bibliotecas instaladas
    installed_packages = pkg_resources.working_set
    python_info['Installed Packages'] = {package.project_name: package.version for package in installed_packages}

    return python_info

def get_system_info():
    system_info = {}
    system_info['Operating System'] = platform.system()
    system_info['OS Version'] = platform.version()
    system_info['Machine Type'] = platform.machine()
    system_info['Platform'] = platform.platform()
    system_info['Node'] = platform.node()
    return system_info

# Extraindo todas as informa√ß√µes
cpu_info = get_cpu_info()
ram_info = get_ram_info()
disk_info = get_disk_info()
gpu_info = get_gpu_info()
python_info = get_python_info()
system_info = get_system_info()

# Exibindo o relat√≥rio completo
print("==== Relat√≥rio do Ambiente de Desenvolvimento Integrado (IDE) ====\n")
print("=== Informa√ß√µes do Sistema ===")
for key, value in system_info.items():
    print(f"{key}: {value}")

print("\n=== Informa√ß√µes do Processador ===")
for key, value in cpu_info.items():
    print(f"{key}: {value}")

print("\n=== Informa√ß√µes de Mem√≥ria RAM ===")
for key, value in ram_info.items():
    print(f"{key}: {value:.2f} GB")

print("\n=== Informa√ß√µes de Armazenamento ===")
for key, value in disk_info.items():
    print(f"{key}: {value:.2f} GB")

print("\n=== Informa√ß√µes de GPU ===")
for key, value in gpu_info.items():
    if isinstance(value, bool):
        print(f"{key}: {'Sim' if value else 'N√£o'}")
    else:
        print(f"{key}: {value}")

print("\n=== Vers√£o do Python e Principais Bibliotecas Instaladas ===")
for key, value in python_info.items():
    if key == 'Installed Packages':
        print(f"\n{key}:")
        for pkg, ver in value.items():
            print(f"  {pkg}: {ver}")
    else:
        print(f"{key}: {value}")
```
