Foi descoberta uma espécie alienígena de ácido ribonucleico (popularmente conhecido como RNA). Os cientistas, por falta de criatividade, batizaram a descoberta de ácido ribonucleico alienígena (RNAA). Similar ao RNA que conhecemos, o RNAA é uma fita composta de várias bases. As bases são B C F S e podem ligar-se em pares. Os únicos pares possíveis são entre as bases B e S e as bases C e F.
Enquanto está ativo, o RNAA dobra vários intervalos da fita sobre si mesma, realizando ligações entre suas bases. Os cientistas perceberam que:

- Quando um intervalo da fita de RNAA se dobra, todas as bases neste intervalo se ligam com suas bases correspondentes;
- Cada base pode se ligar a apenas uma outra base;
- As dobras ocorrem de forma a maximizar o número de ligações feitas sobre fitas;

As figuras abaixo ilustram dobras e ligacões feitas sobre fitas.


image.png




A tarefa será, dada a descrição de uma tira de RNAA, determinar quantas ligações serão realizadas entre suas bases se a tira ficar ativa.

Entrada:

A entrada é composta por diversos casos de teste e termina com EOF. Cada caso de teste possui uma linha descrevendo a sequência de bases da fita de RNAA. Uma fita de RNAA na entrada contém pelo menos 1 e no máximo 300 bases. Não existem espaços entre bases de uma fita da entrada. As bases são 'B', 'C', 'F' e 'S'.

Saída:

Para cada instância imprima uma linha contendo o número total de ligações que ocorre quando a fita descrita é ativada.

```python
def contar_ligacoes(sequencia):
    # Contadores para as bases B e C
    count_B = 0
    count_C = 0
    ligacoes = 0

    # Contabiliza a quantidade de cada base e conta as ligações
    for base in sequencia:
        if base == 'B':
            count_B += 1  # B pode se ligar com S
        elif base == 'S':
            if count_B > 0:  # Se houver um B disponível
                ligacoes += 1
                count_B -= 1  # Remove um B pareado com S
        elif base == 'C':
            count_C += 1  # C pode se ligar com F
        elif base == 'F':
            if count_C > 0:  # Se houver um C disponível
                ligacoes += 1
                count_C -= 1  # Remove um C pareado com F

    return ligacoes

def processar_casos():
    print("Insira as sequências de bases do RNAA (uma por linha, termine com EOF):")
    try:
        while True:
            entrada = input().strip()  # Lê a entrada
            if not entrada:  # Encerra se a entrada for uma linha vazia
                break
            ligacoes = contar_ligacoes(entrada)
            print(ligacoes)  # Exibe apenas o número de ligações
    except EOFError:  # Captura o EOF para encerrar o loop
        print("Obrigado e Bons Estudos")  # Mensagem final

# Chama a função para processar os casos
processar_casos()
```

```python
def contar_ligacoes(sequencia):
    """Conta o número máximo de ligações em uma sequência de RNAA.

    Args:
        sequencia: Uma string representando a sequência de bases.

    Returns:
        O número máximo de ligações.
    """

    pilha = []
    pares = 0
    pares_complementares = {'B': 'S', 'S': 'B', 'C': 'F', 'F': 'C'}

    for base in sequencia:
        if base in pares_complementares:
            if pilha and pilha[-1] == pares_complementares[base]:
                pilha.pop()
                pares += 1
            else:
                pilha.append(base)

    return pares

if __name__ == "__main__":
    while True:
        try:
            sequencia = input()
            if not sequencia:
                break
            resultado = contar_ligacoes(sequencia)
            print(resultado)
        except EOFError:
            break
```

```python
def infix_to_postfix(expression):
    stack = []
    output = ""
    precedence = {'^': 6, '*': 5, '/': 5, '+': 4, '-': 4, '>': 3, '<': 3, '=': 3, '#': 3, '.': 2, '|': 1}

    for char in expression:
        if char.isalpha() or char.isdigit():
            output += char
        elif char == '(':
            stack.append(char)
        elif char == ')':
            while stack and stack[-1] != '(':
                output += stack.pop()
            if not stack:
                return "Syntax Error!"
            stack.pop()
        else:
            while stack and precedence[char] <= precedence.get(stack[-1], 0):
                output += stack.pop()
            stack.append(char)

    while stack:
        output += stack.pop()
    return output

def is_valid(expression):
    for char in expression:
        if not char.isalnum() and char not in '()^*/+-<>=#|.':
            return "Lexical Error!"
    postfix = infix_to_postfix(expression)
    return postfix if postfix != "Syntax Error!" else postfix

if __name__ == "__main__":
    while True:
        try:
            expression = input()
            if not expression:
                break
            result = is_valid(expression)
            print(result)
        except EOFError:
            break
```
