#Exercício 1-d

##Neste algoritmo foi feito a plotagem das curvas referentes a ordenação assintótica para comparação e entendimento do comportamento destas funçãoes no conceito do Big (O), para o caso de algoritmos que possam ser escritos escritos com complexidade nestas funções.

```python
print(86*'*')
print('Algoritmo que Plota o Gráfico Comparativo do Comportamento Assintótico de Sete Funções') #Pequeno cabeçalho
print(86*'*')
print()

import numpy as np                    #Instanciando biblioteca numpy
import matplotlib.pyplot as plt       #Biblioteca matplotlib
from scipy.special import factorial   #Fatorial da biblioteca scipy

# Definindo o intervalo de n
n = np.linspace(1, 10, 1000)  # Usando valores até 10 para evitar crescimento excessivo

# Definindo as funções
f1 = np.ones_like(n)  # f1(n) = 1
f2 = 4 * n  # f2(n) = 4n
f3 = 2 * n * np.log(n)  # f3(n) = 2n log n
f4 = factorial(np.log(n))  # f4(n) = (log n)!
f5 = n * 2**n  # f5(n) = n * 2^n
f6 = 2**(2 * n)  # f6(n) = 2^2n
f7 = factorial(n + 1)  # f7(n) = (n + 1)!

# Plotando as funções
plt.figure(figsize=(12, 8))
plt.plot(n, f1, label=r'$1$', linewidth=2)           #As 7 funções ordenadas
plt.plot(n, f2, label=r'$4n$', linewidth=2)
plt.plot(n, f3, label=r'$2n\log n$', linewidth=2)    #OBS: Já estão ordenadas em ordem crescente
plt.plot(n, f4, label=r'$(\log n)!$', linewidth=2)
plt.plot(n, f5, label=r'$n \cdot 2^n$', linewidth=2)
plt.plot(n, f6, label=r'$2^{2n}$', linewidth=2)
plt.plot(n, f7, label=r'$(n+1)!$', linewidth=2)

# Configurações do gráfico
plt.yscale('log')  # Escala logarítmica no eixo y para visualizar melhor
plt.xscale('log')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('Gráfico Comparativo Assintótico das 7 Funções')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()

print('FIM DO PROGRAMA')   #OBS: Lembrando que o algoritmo não está ordenando de forma crescente as funções e sim somente plotando as curvas do seu comportamento.
```

##Neste algoritmo traça as curvas de duas funções permitindo observar o comportamento assintótico das funções estudadas.

```python
print(86*'*')
print('Algoritmo que Plota o Gráfico Comparativo do Comportamento Assintótico de Duas Funções')  #Cabeçalho
print(86*'*')
print()

import numpy as np               #Bibliotecas como feito no algoritmo acima
import matplotlib.pyplot as plt

# Definindo as funções
def g(n):
    return 5 * n**2 + n

def f(n):
    return 6 * n**2

# Encontrando valores para c e m
# Escolher c e m inicialmente e ajustar conforme necessário
c = 6  # Uma escolha inicial para c
m = 1  # Começando com m = 1

# Gerando valores de n
n_values = np.linspace(1, 100, 400)

# Calculando os valores das funções
g_values = g(n_values)
f_values = c * f(n_values)

# Plotando as funções
plt.figure(figsize=(10, 6))
plt.plot(n_values, g_values, label='g(n) = 5n^2 + n', color='blue')
plt.plot(n_values, f_values, label='cf(n) = 6n^2', color='red', linestyle='--')
plt.xlabel('n')
plt.ylabel('Valores de g(n) e cf(n)')
plt.title('Gráfico Comparativo Assintótico entre g(n) e cf(n)')
plt.legend()
plt.grid(True)
plt.show()

# Verificação se g(n) ≤ c * f(n) para n ≥ m
n_test = 1000  # Testando para um valor grande de n
g_test = g(n_test)
f_test = c * f(n_test)

g_test, f_test, g_test <= f_test

#OBS: Lembrando que o algoritmo não está ordenando de forma crescente as funções e sim somente plotando as curvas do seu comportamento.
```
