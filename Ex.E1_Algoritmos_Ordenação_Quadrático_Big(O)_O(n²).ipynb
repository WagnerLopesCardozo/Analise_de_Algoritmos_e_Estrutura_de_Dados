```python
import random
import time
import matplotlib.pyplot as plt

# Função para realizar o Selection Sort e medir o tempo de execução
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# Função para medir o tempo de execução do Selection Sort e plotar o gráfico de complexidade
def plot_complexity(arr):
    times = []
    n_values = range(1, len(arr) + 1)

    for i in n_values:
        temp_arr = arr[:i]
        start_time = time.time()
        selection_sort(temp_arr)
        end_time = time.time()
        times.append(end_time - start_time)

    # Plotar o gráfico de complexidade
    plt.figure(figsize=(10, 6))
    plt.plot(n_values, times, label='Selection Sort Time Complexity')
    plt.xlabel('Número de Elementos')
    plt.ylabel('Tempo de Execução (segundos)')
    plt.title('Complexidade de Tempo do Selection Sort')
    plt.legend()
    plt.grid(True)
    plt.show()

# Função para exibir o vetor com quebra de linha a cada 45 elementos
def print_vector_with_break(vector, break_point=45):
    for i in range(0, len(vector), break_point):
        print(', '.join(map(str, vector[i:i+break_point])))

# Solicitar ao usuário o número de elementos do vetor
n = int(input("Digite o número de elementos do vetor: "))

# Gerar um vetor com 'n' números aleatórios
arr = [random.randint(1, 100) for _ in range(n)]

# Mostrar o vetor original com quebra de linha
print("Vetor original:")
print_vector_with_break(arr)

# Medir o tempo de execução do Selection Sort
start_time = time.time()
sorted_arr = selection_sort(arr.copy())
end_time = time.time()

# Mostrar o vetor ordenado com quebra de linha
print("\nVetor ordenado:")
print_vector_with_break(sorted_arr)

# Mostrar o tempo gasto para a ordenação
print(f"\nTempo gasto para ordenar: {end_time - start_time:.6f} segundos")

# Mostrar a complexidade assintótica do algoritmo
print("Complexidade assintótica (Big O): O(n^2)")

# Plotar a complexidade da execução
plot_complexity(arr)
```

```python
import random
import time
import matplotlib.pyplot as plt

# Função para realizar o Insertion Sort e medir o tempo de execução
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Função para medir o tempo de execução do Insertion Sort e plotar o gráfico de complexidade
def plot_complexity(arr):
    times = []
    n_values = range(1, len(arr) + 1)

    for i in n_values:
        temp_arr = arr[:i]
        start_time = time.time()
        insertion_sort(temp_arr)
        end_time = time.time()
        times.append(end_time - start_time)

    # Plotar o gráfico de complexidade
    plt.figure(figsize=(10, 6))
    plt.plot(n_values, times, label='Insertion Sort Time Complexity')
    plt.xlabel('Número de Elementos')
    plt.ylabel('Tempo de Execução (segundos)')
    plt.title('Complexidade de Tempo do Insertion Sort')
    plt.legend()
    plt.grid(True)
    plt.show()

# Função para exibir o vetor com quebra de linha a cada 45 elementos
def print_vector_with_break(vector, break_point=45):
    for i in range(0, len(vector), break_point):
        print(', '.join(map(str, vector[i:i+break_point])))

# Solicitar ao usuário o número de elementos do vetor
n = int(input("Digite o número de elementos do vetor: "))

# Gerar um vetor com 'n' números aleatórios
arr = [random.randint(1, 100) for _ in range(n)]

# Mostrar o vetor original com quebra de linha
print("Vetor original:")
print_vector_with_break(arr)

# Medir o tempo de execução do Insertion Sort
start_time = time.time()
sorted_arr = insertion_sort(arr.copy())
end_time = time.time()

# Mostrar o vetor ordenado com quebra de linha
print("\nVetor ordenado:")
print_vector_with_break(sorted_arr)

# Mostrar o tempo gasto para a ordenação
print(f"\nTempo gasto para ordenar: {end_time - start_time:.6f} segundos")

# Mostrar a complexidade assintótica do algoritmo
print("Complexidade assintótica (Big O): O(n^2)")

# Plotar a complexidade da execução
plot_complexity(arr)
```

```python
import random
import time
import matplotlib.pyplot as plt

# Função para realizar o Bubble Sort e medir o tempo de execução
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

# Função para medir o tempo de execução do Bubble Sort e plotar o gráfico de complexidade
def plot_complexity(arr):
    times = []
    n_values = range(1, len(arr) + 1)

    for i in n_values:
        temp_arr = arr[:i]
        start_time = time.time()
        bubble_sort(temp_arr)
        end_time = time.time()
        times.append(end_time - start_time)

    # Plotar o gráfico de complexidade
    plt.figure(figsize=(10, 6))
    plt.plot(n_values, times, label='Bubble Sort Time Complexity')
    plt.xlabel('Número de Elementos')
    plt.ylabel('Tempo de Execução (segundos)')
    plt.title('Complexidade de Tempo do Bubble Sort')
    plt.legend()
    plt.grid(True)
    plt.show()

# Função para exibir o vetor com quebra de linha a cada 45 elementos
def print_vector_with_break(vector, break_point=45):
    for i in range(0, len(vector), break_point):
        print(', '.join(map(str, vector[i:i+break_point])))

# Solicitar ao usuário o número de elementos do vetor
n = int(input("Digite o número de elementos do vetor: "))

# Gerar um vetor com 'n' números aleatórios
arr = [random.randint(1, 100) for _ in range(n)]

# Mostrar o vetor original com quebra de linha
print("Vetor original:")
print_vector_with_break(arr)

# Medir o tempo de execução do Bubble Sort
start_time = time.time()
sorted_arr = bubble_sort(arr.copy())
end_time = time.time()

# Mostrar o vetor ordenado com quebra de linha
print("\nVetor ordenado:")
print_vector_with_break(sorted_arr)

# Mostrar o tempo gasto para a ordenação
print(f"\nTempo gasto para ordenar: {end_time - start_time:.6f} segundos")

# Mostrar a complexidade assintótica do algoritmo
print("Complexidade assintótica (Big O): O(n^2)")

# Plotar a complexidade da execução
plot_complexity(arr)
```

```python
import random
import time
import matplotlib.pyplot as plt

# Função para realizar o Shell Sort e medir o tempo de execução
def shell_sort(arr):
    n = len(arr)
    gap = n // 2

    while gap > 0:
        for i in range(gap, n):
            temp = arr[i]
            j = i
            while j >= gap and arr[j - gap] > temp:
                arr[j] = arr[j - gap]
                j -= gap
            arr[j] = temp
        gap //= 2

    return arr

# Função para medir o tempo de execução do Shell Sort e plotar o gráfico de complexidade
def plot_complexity(arr):
    times = []
    n_values = range(1, len(arr) + 1)

    for i in n_values:
        temp_arr = arr[:i]
        start_time = time.time()
        shell_sort(temp_arr)
        end_time = time.time()
        times.append(end_time - start_time)

    # Plotar o gráfico de complexidade
    plt.figure(figsize=(10, 6))
    plt.plot(n_values, times, label='Shell Sort Time Complexity')
    plt.xlabel('Número de Elementos')
    plt.ylabel('Tempo de Execução (segundos)')
    plt.title('Complexidade de Tempo do Shell Sort')
    plt.legend()
    plt.grid(True)
    plt.show()

# Função para exibir o vetor com quebra de linha a cada 45 elementos
def print_vector_with_break(vector, break_point=45):
    for i in range(0, len(vector), break_point):
        print(', '.join(map(str, vector[i:i+break_point])))

# Solicitar ao usuário o número de elementos do vetor
n = int(input("Digite o número de elementos do vetor: "))

# Gerar um vetor com 'n' números aleatórios
arr = [random.randint(1, 100) for _ in range(n)]

# Mostrar o vetor original com quebra de linha
print("Vetor original:")
print_vector_with_break(arr)

# Medir o tempo de execução do Shell Sort
start_time = time.time()
sorted_arr = shell_sort(arr.copy())
end_time = time.time()

# Mostrar o vetor ordenado com quebra de linha
print("\nVetor ordenado:")
print_vector_with_break(sorted_arr)

# Mostrar o tempo gasto para a ordenação
print(f"\nTempo gasto para ordenar: {end_time - start_time:.6f} segundos")

# Mostrar a complexidade assintótica do algoritmo
print("Complexidade assintótica (Big O): Depende da escolha das lacunas, mas pode ser O(n^2)")

# Plotar a complexidade da execução
plot_complexity(arr)
```

```python
import random
import time
import matplotlib.pyplot as plt

# Função para realizar o Shaker Sort (Cocktail Sort) e medir o tempo de execução
def shaker_sort(arr):
    n = len(arr)
    swapped = True
    start = 0
    end = n - 1
    while swapped:
        swapped = False

        # Passagem da esquerda para a direita
        for i in range(start, end):
            if arr[i] > arr[i + 1]:
                arr[i], arr[i + 1] = arr[i + 1], arr[i]
                swapped = True

        if not swapped:
            break

        swapped = False
        end -= 1

        # Passagem da direita para a esquerda
        for i in range(end - 1, start - 1, -1):
            if arr[i] > arr[i + 1]:
                arr[i], arr[i + 1] = arr[i + 1], arr[i]
                swapped = True

        start += 1

    return arr

# Função para medir o tempo de execução do Shaker Sort e plotar o gráfico de complexidade
def plot_complexity(arr):
    times = []
    n_values = range(1, len(arr) + 1)

    for i in n_values:
        temp_arr = arr[:i]
        start_time = time.time()
        shaker_sort(temp_arr)
        end_time = time.time()
        times.append(end_time - start_time)

    # Plotar o gráfico de complexidade
    plt.figure(figsize=(10, 6))
    plt.plot(n_values, times, label='Shaker Sort Time Complexity')
    plt.xlabel('Número de Elementos')
    plt.ylabel('Tempo de Execução (segundos)')
    plt.title('Complexidade de Tempo do Shaker Sort')
    plt.legend()
    plt.grid(True)
    plt.show()

# Função para exibir o vetor com quebra de linha a cada 45 elementos
def print_vector_with_break(vector, break_point=45):
    for i in range(0, len(vector), break_point):
        print(', '.join(map(str, vector[i:i+break_point])))

# Solicitar ao usuário o número de elementos do vetor
n = int(input("Digite o número de elementos do vetor: "))

# Gerar um vetor com 'n' números aleatórios
arr = [random.randint(1, 100) for _ in range(n)]

# Mostrar o vetor original com quebra de linha
print("Vetor original:")
print_vector_with_break(arr)

# Medir o tempo de execução do Shaker Sort
start_time = time.time()
sorted_arr = shaker_sort(arr.copy())
end_time = time.time()

# Mostrar o vetor ordenado com quebra de linha
print("\nVetor ordenado:")
print_vector_with_break(sorted_arr)

# Mostrar o tempo gasto para a ordenação
print(f"\nTempo gasto para ordenar: {end_time - start_time:.6f} segundos")

# Mostrar a complexidade assintótica do algoritmo
print("Complexidade assintótica (Big O): O(n^2)")

# Plotar a complexidade da execução
plot_complexity(arr)
```

```python
import random
import time
import matplotlib.pyplot as plt

# Função para realizar o Gnome Sort e medir o tempo de execução
def gnome_sort(arr):
    n = len(arr)
    index = 0
    while index < n:
        if index == 0 or arr[index] >= arr[index - 1]:
            index += 1
        else:
            arr[index], arr[index - 1] = arr[index - 1], arr[index]
            index -= 1
    return arr

# Função para medir o tempo de execução do Gnome Sort e plotar o gráfico de complexidade
def plot_complexity(arr):
    times = []
    n_values = range(1, len(arr) + 1)

    for i in n_values:
        temp_arr = arr[:i]
        start_time = time.time()
        gnome_sort(temp_arr)
        end_time = time.time()
        times.append(end_time - start_time)

    # Plotar o gráfico de complexidade
    plt.figure(figsize=(10, 6))
    plt.plot(n_values, times, label='Gnome Sort Time Complexity')
    plt.xlabel('Número de Elementos')
    plt.ylabel('Tempo de Execução (segundos)')
    plt.title('Complexidade de Tempo do Gnome Sort')
    plt.legend()
    plt.grid(True)
    plt.show()

# Função para exibir o vetor com quebra de linha a cada 45 elementos
def print_vector_with_break(vector, break_point=45):
    for i in range(0, len(vector), break_point):
        print(', '.join(map(str, vector[i:i+break_point])))

# Solicitar ao usuário o número de elementos do vetor
n = int(input("Digite o número de elementos do vetor: "))

# Gerar um vetor com 'n' números aleatórios
arr = [random.randint(1, 100) for _ in range(n)]

# Mostrar o vetor original com quebra de linha
print("Vetor original:")
print_vector_with_break(arr)

# Medir o tempo de execução do Gnome Sort
start_time = time.time()
sorted_arr = gnome_sort(arr.copy())
end_time = time.time()

# Mostrar o vetor ordenado com quebra de linha
print("\nVetor ordenado:")
print_vector_with_break(sorted_arr)

# Mostrar o tempo gasto para a ordenação
print(f"\nTempo gasto para ordenar: {end_time - start_time:.6f} segundos")

# Mostrar a complexidade assintótica do algoritmo
print("Complexidade assintótica (Big O): O(n^2)")

# Plotar a complexidade da execução
plot_complexity(arr)
```
